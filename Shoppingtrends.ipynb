```python
def histo(x):
  plt.figure(figsize = (15 , 5))
  print(df[x].plot(kind = 'hist') )
  plt.title(f'\n{x} Distribution\n' , fontsize = 14 , fontweight = 'bold')
  plt.xlabel(f'\n{x}\n' , fontsize = 14)
  plt.ylabel('\nFrequency\n' , fontsize = 14)
  plt.show()
```

```python
def Text(y):
  print(df[y].unique())
  print('===============================================================')
  print(df[y].value_counts().head())
  print('===============================================================')
  plt.figure(figsize = (15 , 5))
  print(df[y].value_counts().head().plot(kind = 'bar') )
  plt.title(f'\n{y} Distribution\n' , fontsize = 14 , fontweight = 'bold')
  plt.xlabel(f'\n{y}\n' , fontsize = 14)
  plt.ylabel('\nFrequency\n' , fontsize = 14)
  plt.xticks(rotation = 0)
  plt.show()
```
