# **Univariate Analysis**

```python
def Text(x):
  print(df[x].nunique())
  print('======================================================')
  print(df[x].value_counts().head())
  print('======================================================')
  print(df[x].value_counts().head().plot(kind = 'bar' , figsize = (15 , 5)))
  plt.title(f'\n{x} Distribution\n' , fontsize = 14 , fontweight = 'bold')
  plt.xlabel(f"\n{x}\n" , fontsize = 12)
  plt.ylabel('\nFrequency\n' , fontsize = 12)
  plt.xticks(rotation = 0)
  plt.show()
```

```python
Text('BusinessTravel')
```

```python
Text('Division')
```

```python
Text('EducationField')
```

```python
def Nums(y):
  # print(df[y].nunique())
  # print('======================================================')
  # print(df[x].value_counts().head())
  print('======================================================')
  print(df[y].plot(kind = 'hist' , figsize = (15 , 5)))
  plt.title(f'\n{y} Distribution\n' , fontsize = 14 , fontweight = 'bold')
  plt.xlabel(f"\n{y}\n" , fontsize = 12)
  plt.ylabel('\nFrequency\n' , fontsize = 12)
  plt.xticks(rotation = 0)
  plt.show()
```

```python
Nums('Daily Wage')
```

```python
Nums('StandardHours')
```

```python
plt.figure(figsize = (15 , 5))
sns.scatterplot(x = 'YearsSinceLastPromotion' , y = 'Daily Wage' , data = df)
plt.title(f'\nScatter Plot of Working Hours Vs Daily Wages\n' , fontsize = 14 , fontweight = 'bold')
plt.xlabel(f"\nWorking Hours\n" , fontsize = 12)
plt.ylabel('\nDaily Wages\n' , fontsize = 12)
plt.xticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (15 , 5))
sns.regplot(x = 'TotalWorkingYears' , y = 'Daily Wage' , data = df)
plt.title(f'\nScatter Plot of Working Hours Vs Daily Wages\n' , fontsize = 14 , fontweight = 'bold')
plt.xlabel(f"\nWorking Hours\n" , fontsize = 12)
plt.ylabel('\nDaily Wages\n' , fontsize = 12)
plt.xticks(rotation = 0)
plt.show()
```

```python
Corrs = df[['DistanceFromHome' , 'Education' , 'Daily Wage' , 'EmployeeCount' , 'StandardHours' , 'TotalWorkingYears']].corr()
```

```python
plt.figure(figsize = (15 , 5))
sns.heatmap( Corrs , annot = True)
plt.title(f'\nCorrelation Detection\n' , fontsize = 14 , fontweight = 'bold')
plt.xlabel(f"\nData Features\n" , fontsize = 12)
plt.ylabel('\nData Features\n' , fontsize = 12)
plt.xticks(rotation = 0)
plt.show()
```
